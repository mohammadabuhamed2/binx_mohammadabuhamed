***2-***

***-Drop the Cabin column because it contains too many missing values and does not significantly impact the study***

***-Fill missing values in the Age column with the average (mean) to preserve the dataset size and avoid data loss***

***-Fill missing values in the Fare column with the median because the dataset contains high-income outliers, making the mean***
 ***inaccurate***

```python
print(titanic.isnull().sum())
titanic.drop(columns="Cabin",inplace=True)
titanic['Age']=titanic['Age'].fillna(titanic['Age'].mean())
titanic['Fare']=titanic['Fare'].fillna(titanic['Fare'].median())
print(titanic.isnull().sum())
```
